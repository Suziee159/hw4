```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("titanic_data.xls")
```

```python
df["Survived"].replace({"Yes": 1, "No": 0}, inplace=True)
```

```python
survived = df[df["Survived"] == 1]["Survived"].count()
not_survived = df[df["Survived"] == 0]["Survived"].count()
```

```python
labels = ["Survived", "Did Not Survive"]
counts = [survived, not_survived]
plt.bar(labels, counts)
plt.xlabel("Outcome")
plt.ylabel("Count")
plt.title("Number of People Who Survived and Did Not Survive")
plt.show()
```
